Будем использовать встроенные средства импьютации библиотеки scikit-learn - https://scikit-learn.org/stable/modules/impute.html#impute

```python
from sklearn.impute import SimpleImputer
from sklearn.impute import MissingIndicator
```

```python
data_cross = data[col]
data_cross
```

С помощью класса SimpleImputer проведем импьютацию с различными показателями центра распределения ("среднее", "медиана", "самое частое")

```python
def test_num_impute_col(dataset, column, strategy_param):
    temp_data = dataset[[column]]
    
    indicator = MissingIndicator()
    mask_missing_values_only = indicator.fit_transform(temp_data)
    
    imp_num = SimpleImputer(strategy = strategy_param)
    data_num_imp = imp_num.fit_transform(temp_data)
    
    filled_data = data_num_imp[mask_missing_values_only]
    
    return column, strategy_param, filled_data.size, filled_data[0], filled_data[filled_data.size-1]
```

```python
# центр распределения - "среднее"
test_num_impute_col(data, col, "mean")
```

```python
# центр распределения - "медиана"
test_num_impute_col(data, col, "median")
```

```python
# центр распределения - "самое частое"
test_num_impute_col(data, col, "most_frequent")
```

Таким образом можно избавиться от пропусков во всех колонках, но это происходит весьма неосознанно

```python
for row in data.itertuples():
     if np.isnan(row.Crossing):
            data_potential = data.loc[data['Potential']== row.Potential]
            data.at[row.Index, col] = data[col].mean()
data.isnull().sum()
```

## 2. Категориальный признак

**Заполнение пропусков в столбце "club"**

```python
# Преобразование типа колонок с пропущенными числовыми значениями в строковый
data['Club'] = data['Club'].astype(str)
```

```python
# Выберем даннее только из этой колонки
col_club = 'Club'
club_data = data[[col_club]]
club_data.head()
```

```python
# Все возможнве клубы (все уникальные значения колонки)
data[col_club].unique()
```

```python
# Размер колонки
data[col_club].unique().shape
```

```python
for row in data.itertuples():
     if pd.isnull(row.Club):
            Nationality_data = data.loc[data['Nationality'] == row.Nationality]
            if (pd.isnull(Nationality_data['Club'].mode(dropna = False)[0])):
                data.at[row.Index, 'Club'] = data['Club'].mode()[0] # Если и это не помогло, то просто находим самое популярное по всей таблице:
            else:
                data.at[row.Index, 'Club'] = Nationality_data['Club'].mode()[0]
data.isnull().sum()
```

Таким образом, мы избавились от пропусков в категориальных данных.
